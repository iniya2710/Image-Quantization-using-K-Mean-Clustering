## Image quantization using K-Means

```python
#Importing the required packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline 
```

```python
#Reading single image as array
img_arr = mpimg.imread('7-colorful-photography.jpg')
```

```python
# Printing the dimension of the image
img_arr.shape 
```

```python
# Visualising the image using matplotlib
plt.imshow(img_arr)
```

```python
#Increasing the visualising size of the image
plt.figure(dpi=130)  #Increase the dimension of pixel 
plt.imshow(img_arr)
```

```python
#Storing the dimension of image seperately as height, width and color channel
H,W,C = img_arr.shape
```

```python
# Comverting 3D array to 2D by multiplying height and width
img_2d = img_arr.reshape(H*W,C)
```

```python
img_2d.shape
```

```python
490*980
```

```python
#Importing k mean clustering algorithm

from sklearn.cluster import KMeans
```

```python
#Building the kmean clustering model
model = KMeans(n_clusters=7)

#Fitting and predicting the model
clust = model.fit_predict(img_2d)
```

```python
clust
```

```python
rgb_centr = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_centr
```

```python
#Converting the predicted 2D to 3D array
pred_3d = rgb_centr[clust].reshape(H,W,C)
```

```python
pred_3d
```

```python
#Visualising the color quantized image
plt.figure(dpi=150)
plt.imshow(pred_3d)
```
